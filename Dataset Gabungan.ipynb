```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
folder_path = '/content/drive/My Drive/dataset/'
```

```python
dataset1 = pd.read_csv(folder_path + 'Recon OS Scan.csv')
```

```python
dataset1
```

```python
dataset2 = pd.read_csv(folder_path + 'DoS UDP Flood.csv')
```

```python
dataset2
```

```python
penggabungan = pd.concat([dataset1, dataset2],ignore_index=True)
```

```python
penggabungan.columns.values
```

```python
x = penggabungan.iloc[:,10: 75]
```

```python
x
```

```python
y = penggabungan.iloc[:,83: 84]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
apip = DecisionTreeClassifier(criterion='entropy', splitter = 'random')
apip.fit(x_train,y_train)
```

```python
y_pred = apip.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
```

```python
accuracy_percent = round(accuracy * 100, 2)
print(f"Akurasi: {accuracy_percent}%")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize = (10, 7))
tree.plot_tree(apip, feature_names = x.columns.values, class_names = np.array([ 'Recon OS Scan','DoS UDP Flood']), filled = True)
plt.show()
```

```python
import seaborn as lol
from sklearn import metrics
label = np.array([ 'Recon OS Scan','DoS UDP Flood'])
```

```python
import matplotlib.pyplot as plt
conf_matrix = metrics.confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 10))
lol.heatmap(conf_matrix, annot=True, xticklabels=label, yticklabels=label)
plt.xlabel('Prediksi')
plt.ylabel('Fakta')
plt.show()
```
